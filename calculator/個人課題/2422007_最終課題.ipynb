```python
# テーブルの作成
def create_table():
    engine = create_engine('sqlite:///jobs.db')
    conn = engine.connect()
    conn.execute('''
        CREATE TABLE IF NOT EXISTS jobs (
            求人番号 TEXT PRIMARY KEY,
            休憩時間 TEXT,
            年間休日数 TEXT,
            休日 TEXT,
            週休二日制 TEXT,
            有給休暇日数 TEXT,
            給与 TEXT,
            基本給 TEXT,
            定額的に支払われる手当 TEXT,
            固定残業代 TEXT,
            月平均労働日数 TEXT,
            通勤手当 TEXT,
            通勤手当月額 TEXT,
            昇給制度 TEXT,
            昇給前年度実績 TEXT,
            昇給金額率 TEXT,
            賞与制度の有無 TEXT,
            賞与金額 TEXT
        )
    ''')
    conn.close()


# 全てのデータを蓄えるリスト
all_job_data = []

# 各ページからデータを抽出
for URL in urls:
    try:
        driver.get(URL)
        time.sleep(2)  # サーバー負荷対策

        # ページのHTMLを取得
        page_content = driver.page_source

        # BeautifulSoupで解析
        soup = BeautifulSoup(page_content, 'html.parser')

        # 求人情報の必要なデータを格納する
        job_data = {
            '求人番号': None,
            '休憩時間': None,
            '年間休日数': None,
            '休日': None,
            '週休二日制': None,
            '有給休暇日数': None,
            '給与': None,
            '基本給': None,
            '定額的に支払われる手当': None,
            '固定残業代': None,
            '月平均労働日数': None,
            '通勤手当': None,
            '通勤手当月額': None,
            '昇給制度': None,
            '昇給前年度実績': None,
            '昇給金額率': None,
            '賞与制度の有無': None,
            '賞与金額': None
        }

        # セレクタに基づくデータの抽出関数
        def extract_by_id(soup, id):
            element = soup.find('div', id=id)
            if element:
                return element.get_text(strip=True)
            return None

        # 各フィールドのデータを抽出
        job_data['求人番号'] = extract_by_id(soup, 'ID_kjNo')
        job_data['休憩時間'] = extract_by_id(soup, 'ID_kyukeiJn')
        job_data['年間休日数'] = extract_by_id(soup, 'ID_nenkanKjsu')
        job_data['休日'] = extract_by_id(soup, 'ID_kyjs')
        job_data['週休二日制'] = extract_by_id(soup, 'ID_shukFtskSei')
        job_data['有給休暇日数'] = extract_by_id(soup, 'ID_nenjiYukyu')
        job_data['給与'] = extract_by_id(soup, 'ID_chgn')
        job_data['基本給'] = extract_by_id(soup, 'ID_khky')
        job_data['定額的に支払われる手当'] = extract_by_id(soup, 'ID_tgktNiShwrTat')
        job_data['固定残業代'] = extract_by_id(soup, 'ID_koteiZngyKbn')
        job_data['月平均労働日数'] = extract_by_id(soup, 'ID_thkinRodoNissu')
        job_data['通勤手当'] = extract_by_id(soup, 'ID_tsknTat')
        job_data['通勤手当月額'] = extract_by_id(soup, 'ID_tsknTatKingaku')
        job_data['昇給制度'] = extract_by_id(soup, 'ID_shokyuSd')
        job_data['昇給前年度実績'] = extract_by_id(soup, 'ID_shokyuMaeNendoJisseki')
        job_data['昇給金額率'] = extract_by_id(soup, 'ID_sokkgSkrt')
        job_data['賞与制度の有無'] = extract_by_id(soup, 'ID_shoyoSdNoUmu')
        job_data['賞与金額'] = extract_by_id(soup, 'ID_shoyoKingaku')

        # 取得したデータをリストに追加
        all_job_data.append(job_data)
    
    except Exception as e:
        print(f"An error occurred: {e}")

# データベースへの保存
try:
    # データフレームに変換
    df = pd.DataFrame(all_job_data)

    # SQLiteデータベースに接続
    engine = create_engine('sqlite:///jobs.db')
    conn = engine.connect()

    # データをテーブルに保存
    df.to_sql('jobs', conn, if_exists='replace', index=False)

    # コネクションを閉じる
    conn.close()

except Exception as e:
    print(f"An error occurred while saving to database: {e}")

finally:
    driver.quit()
```

```python
# データ読込とFletアプリケーションの定義
def main(page):
    # データベースからデータを読み込み
    engine = create_engine('sqlite:///jobs.db')
    conn = engine.connect()
    df = pd.read_sql('SELECT * FROM jobs', conn)
    conn.close()
    job_numbers = df['求人番号'].tolist()

    # 選択された求人番号に基づいて詳細を表示
    def show_job_details(e):
        selected_job = e.control.value
        job_details = df[df['求人番号'] == selected_job].iloc[0]  # データフレームの行を取得

        # テキストコントロールに詳細を表示
        details_text.value = (
            f"求人番号: {job_details['求人番号']}\n"
            f"休憩時間: {job_details['休憩時間']}\n"
            f"年間休日数: {job_details['年間休日数']}\n"
            f"休日: {job_details['休日']}\n"
            f"週休二日制: {job_details['週休二日制']}\n"
            f"有給休暇日数: {job_details['有給休暇日数']}\n"
            f"給与: {job_details['給与']}\n"
            f"基本給: {job_details['基本給']}\n"
            f"定額的に支払われる手当: {job_details['定額的に支払われる手当']}\n"
            f"固定残業代: {job_details['固定残業代']}\n"
            f"月平均労働日数: {job_details['月平均労働日数']}\n"
            f"通勤手当: {job_details['通勤手当']}\n"
            f"通勤手当月額: {job_details['通勤手当月額']}\n"
            f"昇給制度: {job_details['昇給制度']}\n"
            f"昇給前年度実績: {job_details['昇給前年度実績']}\n"
            f"昇給金額率: {job_details['昇給金額率']}\n"
            f"賞与制度の有無: {job_details['賞与制度の有無']}\n"
            f"賞与金額: {job_details['賞与金額']}"
        )
        page.update()

    # ドロップダウンコントロールを作成
    dropdown = ft.Dropdown(
        label="求人番号を選択",
        options=[ft.dropdown.Option(text=job, key=job) for job in job_numbers],
        on_change=show_job_details
    )

    # テキストコントロールを作成
    details_text = ft.Text()

    # ページにコントロールを追加
    page.add(dropdown, details_text)

# Fletアプリケーションの実行
ft.app(target=main)
```
